```python
import numpy as np

# Definição da classe VetorOrdenado
class VetorOrdenado:

    # Método construtor: cria o vetor com a capacidade informada
    def __init__(self, capacidade):
        self.capacidade = capacidade             # capacidade máxima do vetor
        self.ultima_posicao = -1                 # índice da última posição ocupada
        self.valores = np.empty(self.capacidade, dtype=int) # cria o vetor vazio

    # Método para imprimir os elementos do vetor
    # Complexidade: O(n), pois percorre todos os elementos inseridos
    def imprime(self):
        if self.ultima_posicao == -1:            # se não há elementos no vetor
            print('O vetor está vazio')
        else:
            # percorre os elementos já inseridos e imprime com o índice
            for i in range(self.ultima_posicao + 1):
                print(i, ' - ', self.valores[i])

    # Método para inserir um valor no vetor mantendo a ordenação
    # Complexidade: O(n), pois pode ter que deslocar elementos
    def insere(self, valor):
        # Verifica se o vetor já atingiu a capacidade máxima
        if self.ultima_posicao == self.capacidade - 1:
            print('Capacidade máxima atingida')
            return

        # Determina a posição correta para inserir o valor
        posicao = 0
        for i in range(self.ultima_posicao + 1):
            posicao = i
            # se encontrar um valor maior, interrompe e insere antes dele
            if self.valores[i] > valor:
                break
            # se chegar no final, insere na última posição + 1
            if i == self.ultima_posicao:
                posicao = i + 1

        # Desloca todos os elementos para a direita para abrir espaço
        x = self.ultima_posicao
        while x >= posicao:
            self.valores[x + 1] = self.valores[x]
            x -= 1

        # Insere o valor na posição correta
        self.valores[posicao] = valor
        # Atualiza a última posição ocupada
        self.ultima_posicao += 1

# Teste do código (parte inferior da imagem)
vetor = VetorOrdenado(10)
vetor.imprime()
```
